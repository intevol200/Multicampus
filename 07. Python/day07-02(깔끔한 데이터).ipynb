## melt 메서드 사용하기

```python
데이터프레임의 열은 열 자체가 값을 의미
데이터프레임의 열이 길게 늘어선 형태를 넓은 데이터라고 한다. 
melt는 데이터를 깔끔하게 정리하는데 유용하다. ( melt : 용해, 녹이다. 차츰차츰변하다. )
```

```python
#pivot_table : 피봇테이블(pivot table)이란 
#데이터 열 중에서 두 개의 열을 각각 행 인덱스, 열 인덱스로 사용하여 데이터를 조회하여 펼쳐놓은 것을 말한다.
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'], 
    columns='element',   
    values='기온'
)
weather_tidy.head(5)  #element의 값이 (tmax,tmin) 칼럼이됨 
```

```python
weather_tidy_flat = weather_tidy.reset_index()   
weather_tidy_flat.index.name = "순서"
weather_tidy_flat.head()
```

## 빌보드 차트의 중복 데이터 처리하기

```python
billboard = pd.read_csv('data/billboard.csv')
billboard_long = pd.melt(billboard, id_vars=['year', 'artist', 'track', 'time', 'date.entered'], 
                         var_name='week', 
                         value_name='rating') #평점 

print(billboard_long.shape)
billboard_long.head()
```

```python
#중복데이터가 많음을 볼 수 있다. (year, artist,  track,  time, date)
billboard_long.artist.sort_values(ascending=False)
billboard_long[billboard_long.artist=='matchbox twenty'].count()
```

```python
#중복데이터를 가지고있는 열을 따로 모아 새로운 데이터프레임 만들기 
billboard_songs = billboard_long[['year', 'artist', 'track', 'time']] 
print(billboard_songs.shape)
billboard_songs.head()
```

```python
#중복데이터제거 
billboard_songs = billboard_songs.drop_duplicates() 
print(billboard_songs.shape)
billboard_songs.head()
```

```python
#id를 추가하여 데이터를 쉽게 조회하거나 구분할 수 있다.  ....0~316
billboard_songs['id'] = range(len(billboard_songs)) 
billboard_songs.head()
```

```python
#노래 정보와 주간 순위 데이터를 합친것이다.....317건 ---24092건 ===>24092건
billboard_ratings = billboard_long.merge( 
    billboard_songs, on=['year', 'artist', 'track', 'time']) 
print(billboard_ratings.shape)
```

```python
billboard_ratings.head()
```

```python
import os 
import urllib.request

# 네트워크 상태에 따라 5 ~ 15분이 소요됩니다.
# txt파일에 저장된 URL주소의 데이터를 다운하여 
with open('data/raw_data_urls.txt', 'r') as data_urls:
    for line, url in enumerate(data_urls):
        url = url[:-1] #데이터 파일 문제 해결(\n 따라오는 문제)
        if line == 5:
            break 
        filename = os.path.basename(url) #파일이름
        filepath = os.path.join('data3', filename) #data폴더 아래에 저장하기 위해 filepath 
        print(url, filepath)
        urllib.request.urlretrieve(url, filepath)
```
