## 앤스콤 데이터 집합 불러온 후 그래프 그리기(77쪽)

### 1. 앤스콤 데이터 집합 불러오기

```python
#데이터 시각화를 보여주는 전형적인 사례 : 앤스콤 4분할 그래프 
#영국의 프랭크 앤스콤이 데이터를 시각화하지 않고 수치만 확인했을때 
#발생할 수 있는 함정을 보여주기 위해 만든 그래프 

#평균,분산,상관관계,회귀선의 통계수치가 같아도 그래프의 형태는 다를수 있다.
##데이터분석 수치에만 의존하면 잘못된 판단을 할 수 있다는 것을 알려주는 좋은 예시이다. 

#matplotlib : mathworks에서 개발한 매트랩(MATLAB)과 비슷한 형태
#  다양한 데이터를 많은 방법으로 도식화 할 수 있도록 하는 파이썬 라이브러리로써, 
#  우리는 matplotlib의 pyplot을 이용하게 됩니다
# seaborn : Matplotlib을 기반으로 다양한 색상 테마와 통계용 차트 등의 기능을 추가한 시각화 패키지이다. 
import seaborn as sns 
import matplotlib.pyplot as plt

adata = sns.load_dataset("anscombe")
```

```python
adata.shape
```

```python
#Ⅰ,Ⅱ,Ⅲ,Ⅳ 4개의 그룹의 평균,분산,상관관계,회귀선의 통계수치는 같다. 
#한번에 4개의 그래프 그리기위해 데이터 추출 ⅠⅡ ⅢⅣ
dataset_1 = adata[adata["dataset"]=="I"]
dataset_2 = adata[adata["dataset"]=="II"]
dataset_3 = adata[adata["dataset"]=="III"]
dataset_4 = adata[adata["dataset"]=="IV"]
#평균
avg1 = dataset_1.mean()
avg2 = dataset_2.mean()
avg3 = dataset_3.mean()
avg4 = dataset_4.mean()
print(avg1, avg2, avg3, avg4)
#분산
var1 = dataset_1.var()
var2 = dataset_2.var()
var3 = dataset_3.var()
var4 = dataset_4.var()
print(var1, var2, var3, var4)
#표준편차
std1 = dataset_1.std()
std2 = dataset_2.std()
std3 = dataset_3.std()
std4 = dataset_4.std()
print(std1, std2, std3, std4)
#상관계수
corr1 = dataset_1["x"].corr(dataset_1["y"])
corr2 = dataset_2["x"].corr(dataset_2["y"])
corr3 = dataset_3["x"].corr(dataset_3["y"])
corr4 = dataset_4["x"].corr(dataset_4["y"])
print(corr1, corr2, corr3, corr4)
```

```python
%matplotlib inline  
#notebook을 실행한 브라우저에서 바로 그림을 볼 수 있게 해주는 것 입니다.
import matplotlib.pyplot as plt
```

```python
#선 그래프 그리기 
plt.plot(dataset_1['x'], dataset_1['y'])
```

```python
#점으로 그래프 그리기 
plt.plot(dataset_1['x'], dataset_1['y'], 'o')
```

```python
#matplotlib 라이브러리로 그래프 그리기 
#1.전체 그래프가 위치할 기본틀을 만든다.
#2.그래프를 넣을 그래프 격자를 만든다.
#3.격자에 그래프를 하나씩 추가한다.(왼쪽->오른쪽방향으로  위->아래방향으로 )
```

```python
fig = plt.figure(figsize=(7,5))

axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)
```

```python
axes1.plot(dataset_1['x'], dataset_1['y'], 'o') 
axes2.plot(dataset_2['x'], dataset_2['y'], 'o') 
axes3.plot(dataset_3['x'], dataset_3['y'], 'o') 
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')

fig
```

```python
axes1.set_title("dataset_1") 
axes2.set_title("dataset_2")
axes3.set_title("dataset_3") 
axes4.set_title("dataset_4")

fig
```
